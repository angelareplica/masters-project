```python
import pandas as pd
import re
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('SEC_lit_releases_2010.csv')
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
df.head()
```

```python
# make a new column of lowercase release numbers to generate URL column
df['release_no'] = df['Release No.'].str.replace('-', '')
df['release_no'] = df['release_no'].str.lower()
```

```python
# add the url column based on the url pattern
df['url'] = "https://www.sec.gov/litigation/litreleases/2010/" + df.release_no + ".htm"
```

```python
# drop the release_no column
df = df.drop(['release_no'], axis=1)
```

```python
df
```

```python
# using BeautifulSoup to loop through each URL and pull all the text

def get_release_info(row):
    
    try:
        request = requests.get(row['url'])
        soup = BeautifulSoup(request.text, 'html.parser')
        
        full_text = soup.findAll(text=True)
        
    except:
        return pd.Series({})

    return pd.Series({
        'full_text': full_text,
    })
```

```python
df_litreleases_2010 = df.apply(get_release_info, axis=1).join(df)
```

```python
df_litreleases_2010.head()
```

```python
df_litreleases_2010.dtypes
```

```python
df_litreleases_2010['Respondents'] = df_litreleases_2010['Action']
```

```python
# rearrange columns
df_litreleases_2010 = df_litreleases_2010[['Release No.','Respondents','Date','url', 'full_text']]
df_litreleases_2010.head()
```

```python
# Save to CSV
df_litreleases_2010.to_csv("litreleases_2010_scraped.csv", index=False)
```
