```python
import pandas as pd
import re
import numpy as np
from sqlalchemy import create_engine
```

```python
df_2009 = pd.read_csv('litreleases_2009_scraped.csv')
df_2010 = pd.read_csv('litreleases_2010_scraped.csv')
df_2011 = pd.read_csv('litreleases_2011_scraped.csv')
df_2012 = pd.read_csv('litreleases_2012_scraped.csv')
df_2013 = pd.read_csv('litreleases_2013_scraped.csv')
df_2014 = pd.read_csv('litreleases_2014_scraped.csv')
df_2015 = pd.read_csv('litreleases_2015_scraped.csv')
df_2016 = pd.read_csv('litreleases_2016_scraped.csv')
df_2017 = pd.read_csv('litreleases_2017_scraped.csv')
df_2018 = pd.read_csv('litreleases_2018_scraped_full.csv')
```

```python
dataframes = [df_2009, df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018]
df = pd.concat(dataframes)
```

```python
# Save to CSV
df.to_csv("litreleases_concatenated_updated2019.csv", index=False)
```
